```python
import pandas as pd
df=pd.read_excel('/content/brainsize.xlsx')
df.head()
```

```python
from numpy.ma.core import shape
print(df)
print(df.describe())
print(df.shape)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X=df['Head Size(cm^3)'].values
Y=df['Brain Weight(grams)'].values
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
mean_x=np.mean(X)
mean_y=np.mean(Y)
m=len(X)
numer=0
denom=0
for i in range(m):
   numer += (X[i]-mean_x)*(Y[i]-mean_y)
   denom +=(X[i]-mean_x)**2
   b1=numer/denom
   b0=mean_y-(b1*mean_x)
```

```python
plt.rcParams['figure.figsize']=(5.0,2.5)
max_x=np.max(X)+ 10
min_X=np.min(X)-1
x=np.linspace(min_X,max_x,10)
y=b0+b1*x
plt.plot(x,y,color='#58b970',label='Regression line')
plt.scatter(X,Y,c='#ef5423',label='Scatter plot')
plt.xlabel('Head Size(cm^3)')
plt.ylabel('Brain Weight(grams)')
plt.legend()
plt.show()
```
